```python
class SavingsAccount(BankAccount):
  def __init__(self,owner,accno,balance,interest_rate):
    super().__init__(owner,accno,balance)
    self.interest_rate = interest_rate

  def calculate_interest(self):
    interest = self.balance * self.interest_rate # Corrected attribute name
    self.balance = self.balance + interest
    print(f"Interest Amount is {interest}")

  def withdraw(self,amount):
    if self.balance <= 2000:
      print("Withdrawal restricted. MinimumBalance should be maintained")
    elif self.balance < amount:
      print("Insufficient Balance")
    else:
      super().withdraw(amount)

sav_acc1 = SavingsAccount("Anand",123456,3000,2.5) # Moved outside class definition
sav_acc1.deposit(2000)
sav_acc1.calculate_interest()
sav_acc1.withdraw(5000)
```
